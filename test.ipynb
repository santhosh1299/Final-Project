```python
import pandas as pd
import numpy as np
```

```python

df  = pd.read_csv("h.csv")

df
```

```python
# Check for infinite values
inf_values = df.isin([np.inf, -np.inf])

# Display rows with infinite values
print("Rows with infinite values:")
print(df[inf_values.any(axis=1)])

# Display columns with infinite values
print("Columns with infinite values:")
print(df.columns[inf_values.any()])
```
